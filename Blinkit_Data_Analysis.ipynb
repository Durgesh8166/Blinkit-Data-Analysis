```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
```

```python
df=pd.read_csv('/content/blinkit_data.csv')
df
```

```python
df.head(10)
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df.columns
```

```python
#check missinng values
df.isnull().sum()
```

```python
#filling missing of Item Weight using median
item_weight_median=df.groupby('Item Type')['Item Weight'].transform('median')
df['Item Weight']=df['Item Weight'].fillna(item_weight_median)
```

```python
df['Item Weight']
```

```python
 #Clean Inconsistent Categories ---
print(df['Item Fat Content'].value_counts())
```

```python
# Standardize values in 'Item Fat Content'
df['Item Fat Content']=df['Item Fat Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

```python
df['Item Fat Content'].unique()
```

```python
#Check Duplicates ---
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
# item types generate the highest sales
item_sales=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
```

```python
df_item_sales = pd.DataFrame({
    'Item Type': item_sales.index,
    'Total Sales': item_sales.values
})

# Create the animated bar plot using Plotly Express
fig = px.bar(df_item_sales,
             x='Total Sales',
             y='Item Type',
             title='Total Sales by Item Type',
             color='Total Sales',
             orientation='h',
             labels={'Total Sales': 'Total Sales', 'Item Type': 'Item Type'},
             range_x=[0, df_item_sales['Total Sales'].max() + 1000])  # Setting the max sales value for proper scaling

# Add text labels on bars
fig.update_traces(text=df_item_sales['Total Sales'].round(1), textposition='inside', textfont=dict(color='black', size=12, family="Arial"))

# Animate by adding a frame (optional for more animation effects)
fig.update_layout(
    xaxis_title='Total Sales',
    yaxis_title='Item Type',
    yaxis_tickangle=45,
    title_text="Total Sales by Item Type (Animated)",

)
```

```python
sales_by_fat_content = df.groupby('Item Fat Content')['Sales'].mean()
sales_by_fat_content
```

```python

# Plot the average sales by fat content
plt.figure(figsize=(8, 6))
sns.barplot(x=sales_by_fat_content.index, y=sales_by_fat_content.values, palette='viridis')
plt.title('Average Sales by Item Fat Content')
plt.xlabel('Item Fat Content')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Ratings vs. Fat Content
ratings_by_fat_content = df.groupby('Item Fat Content')['Rating'].mean()
ratings_by_fat_content
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=ratings_by_fat_content.index, y=ratings_by_fat_content.values, palette='viridis')
plt.title('Average Ratings by Item Fat Content')
plt.xlabel('Item Fat Content')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Item Fat Content' and calculate the mean of both 'Sales' and 'Rating'
fat_content_analysis = df.groupby('Item Fat Content')[['Sales', 'Rating']].mean()

# Plot the results
fat_content_analysis.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Sales and Ratings by Fat Content')
plt.ylabel('Values')
plt.xlabel('Fat Content')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#combinations of item type and outlet type generate the most revenue
revenue_by_combination = df.groupby(['Item Type', 'Outlet Type'])['Sales'].sum().reset_index()
revenue_by_combination = revenue_by_combination.sort_values('Sales', ascending=False)
revenue_by_combination
```

```python
fig = px.bar(revenue_by_combination,
             x="Item Type",
             y="Sales",
             color="Outlet Type",
             barmode="group",
             title="Sales by Item Type and Outlet Type",
             height=600)
fig.update_layout(xaxis_tickangle=-40)
```

```python
#the characteristics of top-performing outlets (high sales + high ratings)
top_performing_outlets =df[(df['Sales']>df['Sales'].quantile(0.75)) & (df['Rating']>df['Rating'].quantile(0.75))]
charterstic_df=top_performing_outlets[['Outlet Size', 'Outlet Type', 'Outlet Location Type']]
charterstic_df
```

```python
charterstic_df = charterstic_df.value_counts().reset_index(name='Count')
```

```python
px.bar(charterstic_df,x='Outlet Type',y='Count',color='Outlet Size',facet_col='Outlet Location Type',barmode='group',

             title='Characteristics of Top-Performing Outlets (High Sales + High Ratings)',
             labels={'Count': 'Number of Outlets'},
             height=500)

```

```python
# Calculate outlet age
current_year=2025
df['Outlet Age']=current_year-df['Outlet Establishment Year']
```

```python
df.sample(5)
```

```python
# Group by age and calculate average sales
age_sales = df.groupby('Outlet Age')['Sales'].mean().reset_index()
age_sales
```

```python

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Outlet Age', y='Sales', data=age_sales, marker='o')
plt.title('Outlet Age vs. Average Sales')
plt.xlabel('Outlet Age (years)')
plt.ylabel('Average Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by Item Type and calculate average rating
item_ratings = df.groupby('Item Type')['Rating'].mean().sort_values(ascending=False).reset_index()
item_ratings
```

```python
fig = px.bar(
    item_ratings,
    x='Rating',
    y='Item Type',
    orientation='h',
    color='Item Type',
    #color_continuous_scale='Blues',
    title='Average Rating by Item Type',
    labels={'Rating': 'Average Rating', 'Item Type': 'Item Type'}
)

fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},  # Highest rating on top
    title_font_size=20,
    xaxis_title='Average Rating',
    yaxis_title='Item Type'
)
```

```python
 #unusually high/low values that might indicate errors or special products.

cols = ['Sales', 'Item Visibility', 'Item Weight']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
    plt.grid(True)

plt.tight_layout()
```

```python
#Top 10 Best-Selling Item Categories
top_items = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
top_items
```

```python
fig = px.bar(
    top_items,
    x='Sales',
    y='Item Type',
    orientation='h',
    title='Top 10 Best-Selling Item Categories',
    color='Sales',
    color_continuous_scale='RdBu',
    labels={'Sales': 'Total Sales', 'Item Type': 'Item Type'}
)

fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},  # keeps highest sales at top
    title_font_size=20,
    xaxis_title='Total Sales',
    yaxis_title='Item Type'
)
```
